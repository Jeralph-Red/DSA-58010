```python
def display_instructions():
    print_title("INSTRUCTIONS")
    print("TO PLAY THE GAME:")
    print("> Choose 1 from the main menu.")
    print("> Press enter to start the game.")
    print("> When a question is shown, choose the number of the correct answer.")
    line_break("-")
    print("TO ADD QUESTIONS:")
    print("> Choose 2 from the main menu.")
    print("> Enter the number of questions you would like to add.")
    print("> For every question, enter the question, choices, and correct answer.")
    print("> Blank inputs are considered invalid.")
    print("> There will be 4 choices per question. No choice must repeat.")
    print("> The correct answer must be in the choices.")


def about_us():
    print_title("About Us")

    print("This program was made for the Final Project in DSA")
    print("Adamson University, First Semester A.Y 2022-2023")
    print("Created by: Group 6")


# allows user to input the new question
def enter_question(num): 
    vaild_question = False
    print(f"Enter question {num}:")

    while not vaild_question:
        question = input("--> ")
        if len(question) > 0:
            vaild_question = True
        else:
            print("Question can't be blank! Enter again.")
    return question


# allows user to enter the choices for new questions
# a choice must not be blank and must only be entered once 
def enter_choices(num):
    line_break("-")
    print(f"Enter choices for question {num}")
    valid_choices = False
    choices = []
    ctr = 1

    while not valid_choices:
        choice = input(f"Choice {ctr}: ")

        if len(choice) > 0:
            if choice.lower() not in [ch.lower() for ch in choices]:
                choices.append(choice)
                ctr += 1
            else:
                print("You have already entered that choice! Try again.")
        else:
            print("Choice can't be blank! Try again")
        
        if len(choices) == 4:
            valid_choices = True

    return choices
    

# allows user to enter the correct answer for the new question
# correct answer must be in the inputted choices
def enter_answer(num, choices):
    line_break("-")
    print(f"Enter correct answer for question {num}:")

    valid_answer = False
    while not valid_answer:
        answer = input("Correct answer: ")
        
        if answer.lower() not in [ch.lower() for ch in choices]:
            print("Error. Answer must be in the choices.")
        else:
            valid_answer = True
    
    return answer


def add_questions(currentQuestions, currentChoices, currentAnswers):
    print_title("Add Questions")

    # ask user for the number of questions to add
    valid_num_questions = False

    while not valid_num_questions:
        print("How many questions would you like to add?")
        print("Type 0 to exit.")
        try:
            num_questions = int(input("--> "))

            if num_questions < 0:
                print("Please enter a valid number.")
                line_break("-")
            else:
                valid_num_questions = True
        except ValueError:
            print("Please enter valid number.")
            line_break("-")

    line_break("=")
    for i in range(num_questions):
        new_question = enter_question(i+1)
        new_choices = enter_choices(i+1)
        new_answer = enter_answer(i+1, new_choices)

        currentQuestions.append(new_question)
        currentChoices.append(new_choices)
        currentAnswers.append(new_answer)
        line_break("=")

    if num_questions > 0:
        print("Questions successfully added!")
    return currentQuestions, currentChoices, currentAnswers

# performs the actual quiz
def start_game(questions, choices, answers):
    print_title("Welcome to Capital Quiz")
    input("Press enter to start...")
    line_break("=")

    score = 0
    for i in range(len(questions)):
        question = questions[i]
        question_choices = choices[i]
        correct_answer = answers[i]

        # display question
        print(f"Question {i+1}:")
        print(question)             
        
        # display choices
        for j in range(len(question_choices)):
            print(f"[{j+1}] {question_choices[j]}")
        
        # get user's answer
        automatic_incorrect = False
        try:
            user_answer = int(input("Number of your answer: "))
        except ValueError:
            automatic_incorrect = True

        line_break("-")

        # check if user's answer is correct
        incorrect_answer = False
        if not automatic_incorrect:
            try:
                answer_index = user_answer - 1
                if (question_choices[answer_index] == correct_answer):
                    score += 1
                    print("Correct answer!")
                else:
                    incorrect_answer = True
            except IndexError:
                incorrect_answer = True

        if incorrect_answer or automatic_incorrect:
            print(f"Incorrect! The correct answer is {correct_answer}.")
        
        line_break("-")
    
    # display final score
    line_break("=")
    print(f"Your final score is {score}.")
    line_break("=")

    # ask user if they want to restart the game
    print("Would you like to play the game again?")
    print("Enter Y if yes, enter any character if no")
    restart = input("--> ")
    line_break("=")

    if restart.upper() == "Y":
        start_game(questions, choices, answers)
    else:
        print("Exiting the game...")


# allows user to select an action from the main menu
def choose_action():
    valid_action = False

    while not valid_action:
        action = input("Select a number: ")
        if action in ["1", "2", "3", "4", "5"]:
            valid_action = True
        else:
            print("Please select from 1-5 only!")
    
    return action


# prints a divider
def line_break(char):
    print(char*40)


# for displaying the title according to the selected option from the menu
def print_title(title):
    line_break("=")
    print(title.upper())
    line_break("=")


# displays the main menu
def print_menu():
    line_break("=")
    print("WECLOME TO QUIZ SYSTEM".center(40))
    line_break("=")
    print("[1] Play Quiz")
    print("[2] Add Quiz Questions")
    print("[3] See Instructions")
    print("[4] About Us")
    print("[5] Exit")
    line_break("=")

# main program - entry point
def main():

    questions = [
        "What is the capital of France?",
        "What is the capital of Germany?",
        "What is the capital of Italy?",
        "What is the capital of Spain?",
        "What is the capital of the United Kingdom?"
        
    ]

    choices = [
        ["Paris", "Bangkok", "San Marino", "Moscow"],
        ["Taipei", "Berlin", "Beijing", "Manila"],
        ["Dubai", "Mellington", "Rome", "Tokyo"],
        ["Bangkok", "Moscow", "Mellington", "Madrid"],
        ["London", "Beijing", "Havana", "Taipei"]
    ]

    answers = [
        "Paris",
        "Berlin",
        "Rome",
        "Madrid",
        "London"
    ]

    done_playing = False
    while not done_playing:
        print_menu()
        action = choose_action()

        if action == "1":
            start_game(questions, choices, answers)
        elif action == "2":
            questions, choices, answers = add_questions(questions, choices, answers)
        elif action == "3":
            display_instructions()
        elif action == "4":
            about_us()
        elif action == "5":
            print_title("Program Exit")
            done_playing = True
           

main()
```
